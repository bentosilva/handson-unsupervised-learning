```
# !pip install gtts
import gtts
#τό πλοΐον έστιν έν Βυζαντίώ.
from gtts import gTTS #Import Google Text to Speech
from IPython.display import Audio #Import Audio method from IPython's Display Class
ttsel= gTTS('τό πλοΐον έστιν έν Βυζαννντίώ.', lang='el')
ttsbr= gTTS('Você precisa repensar os conceitos, porque eles estão equivocaaaaaados na sua cabeça. Corrigindo o que você está entendendo por estes conceitos', lang='pt') #Provide the string to convert to speech
ttsel.save('el.wav') #save the string converted to speech as a .wav file
el = 'el.wav'
ttsbr.save('br.wav')
br = 'br.wav'
Audio(el, autoplay=True)
# gtts.lang.tts_langs()
```

```
import soundfile as sf # Sound files can be read or written directly using the functions read() and write()
import IPython.display as ipd # Using IPython.display.Audio, you can play an audio file
import numpy # Audio can accept a NumPy array. Let's synthesize a pure tone at 440 Hz
sr = 22050 # sample rate
T = 2.5    # seconds
t = numpy.linspace(0, T, int(T*sr), endpoint=False) # time variable
x = 0.5*numpy.sin(2*numpy.pi*440*t)                 # pure sine wave at 440 Hz
ipd.Audio(x, rate=sr)                               # load a NumPy array
```
